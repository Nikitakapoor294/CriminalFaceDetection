```
import sqlite3
import os

DB_PATH = 'database/criminals.db'

def get_connection():
    conn = sqlite3.connect(DB_PATH)
    return conn

def init_db():
    conn = get_connection()
    cursor = conn.cursor()


    cursor.execute('''
        CREATE TABLE IF NOT EXISTS criminals (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            name TEXT NOT NULL,
            age INTEGER,
            crime_type TEXT,
            status TEXT DEFAULT 'wanted',
            photo_path TEXT,
            face_encoding BLOB,
            registered_at DATETIME DEFAULT CURRENT_TIMESTAMP
        )
    ''')

    cursor.execute('''
        CREATE TABLE IF NOT EXISTS detection_logs (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            criminal_id INTEGER,
            confidence_score REAL,
            detected_at DATETIME DEFAULT CURRENT_TIMESTAMP,
            image_snapshot BLOB,
            FOREIGN KEY (criminal_id) REFERENCES criminals(id)
        )
    ''')

    cursor.execute('''
        CREATE TABLE IF NOT EXISTS users (
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            username TEXT UNIQUE NOT NULL,
            password_hash TEXT NOT NULL,
            role TEXT DEFAULT 'viewer'
        )
    ''')

    conn.commit()
    conn.close()
    print("✅ Database initialized successfully!")

init_db()

```

```
import face_recognition
import pickle

def encode_face(image_path):
    image = face_recognition.load_image_file(image_path)
    encodings = face_recognition.face_encodings(image)
    if len(encodings) == 0:
        print(" No face found in image.")
        return None
    return pickle.dumps(encodings[0])

def decode_face(blob):
    return pickle.loads(blob)

def compare_faces(known_blob, unknown_encoding, tolerance=0.5):
    known_encoding = decode_face(known_blob)
    distance = face_recognition.face_distance([known_encoding], unknown_encoding)[0]
    match = distance <= tolerance
    confidence = round((1 - distance) * 100, 2)
    return match, confidence

print("Face encoder ready!")
```

```
import shutil

def add_criminal(name, age, crime_type, status, photo_path):
    encoding_blob = encode_face(photo_path)
    if encoding_blob is None:
        return False, "No face detected in photo."

    filename = f"{name.replace(' ', '_')}_{os.path.basename(photo_path)}"
    dest = os.path.join('photos', filename)
    shutil.copy(photo_path, dest)

    conn = get_connection()
    cursor = conn.cursor()
    cursor.execute('''
        INSERT INTO criminals (name, age, crime_type, status, photo_path, face_encoding)
        VALUES (?, ?, ?, ?, ?, ?)
    ''', (name, age, crime_type, status, dest, encoding_blob))
    conn.commit()
    conn.close()
    return True, "Criminal record added successfully!"

def get_all_criminals():
    conn = get_connection()
    cursor = conn.cursor()
    cursor.execute("SELECT id, name, age, crime_type, status FROM criminals")
    records = cursor.fetchall()
    conn.close()
    return records

def search_criminals(keyword):
    conn = get_connection()
    cursor = conn.cursor()
    cursor.execute('''
        SELECT id, name, age, crime_type, status FROM criminals
        WHERE name LIKE ? OR crime_type LIKE ?
    ''', (f'%{keyword}%', f'%{keyword}%'))
    results = cursor.fetchall()
    conn.close()
    return results

def delete_criminal(criminal_id):
    conn = get_connection()
    cursor = conn.cursor()
    cursor.execute("DELETE FROM criminals WHERE id=?", (criminal_id,))
    conn.commit()
    conn.close()
    return "Record deleted."

print("Criminal manager ready!")
```

```
def load_all_encodings():
    conn = get_connection()
    cursor = conn.cursor()
    cursor.execute("SELECT id, name, face_encoding FROM criminals WHERE face_encoding IS NOT NULL")
    records = cursor.fetchall()
    conn.close()
    return records

def log_detection(criminal_id, confidence):
    conn = get_connection()
    cursor = conn.cursor()
    cursor.execute('''
        INSERT INTO detection_logs (criminal_id, confidence_score)
        VALUES (?, ?)
    ''', (criminal_id, confidence))
    conn.commit()
    conn.close()

def scan_image(image_path):
    image = face_recognition.load_image_file(image_path)
    unknown_encodings = face_recognition.face_encodings(image)

    if not unknown_encodings:
        print("No face detected in image.")
        return

    records = load_all_encodings()
    if not records:
        print("No criminals in database yet.")
        return

    for unknown_enc in unknown_encodings:
        best_match = None
        best_conf = 0

        for (cid, cname, cblob) in records:
            match, confidence = compare_faces(cblob, unknown_enc)
            if match and confidence > best_conf:
                best_conf = confidence
                best_match = {"id": cid, "name": cname, "confidence": confidence}

        if best_match:
            print(f" MATCH FOUND: {best_match['name']} | Confidence: {best_match['confidence']}%")
            log_detection(best_match['id'], best_match['confidence'])
        else:
            print("No criminal match found.")

print("Detector ready!")
```

```
from google.colab import files
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np

print("Upload a photo to scan for criminal match...")
uploaded = files.upload()

if uploaded:
    filename = list(uploaded.keys())[0]

    # Save the file
    with open(filename, 'wb') as f:
        f.write(uploaded[filename])

    # Scan the image
    print("\nScanning image for criminal match...")
    scan_image(filename)

    # Show the image with result
    img = Image.open(filename)
    image_array = np.array(img)

    # Draw face locations
    face_locations = face_recognition.face_locations(image_array)

    fig, ax = plt.subplots(1)
    ax.imshow(img)

    for (top, right, bottom, left) in face_locations:
        rect = patches.Rectangle(
            (left, top),
            right - left,
            bottom - top,
            linewidth=2,
            edgecolor='red',
            facecolor='none'
        )
        ax.add_patch(rect)

    plt.title("Face Detection Result")
    plt.axis('off')
    plt.show()

    print(f"\nTotal faces detected: {len(face_locations)}")
else:
    print("No file uploaded. Please try again.")
```

```
def display_all_criminals():
    records = get_all_criminals()
    if not records:
        print("No criminals in database.")
        return

    print("\n" + "=" * 70)
    print("              ALL CRIMINAL RECORDS")
    print("=" * 70)
    print(f"{'ID':<5} {'Name':<20} {'Age':<5} {'Crime':<20} {'Status':<10}")
    print("-" * 70)

    for r in records:
        print(f"{r[0]:<5} {r[1]:<20} {r[2]:<5} {r[3]:<20} {r[4]:<10}")

    print("-" * 70)
    print(f"Total records: {len(records)}")
    print("=" * 70)

def display_search_results(keyword):
    results = search_criminals(keyword)
    if not results:
        print(f"No criminals found for keyword: {keyword}")
        return

    print("\n" + "=" * 70)
    print(f"         SEARCH RESULTS FOR: {keyword.upper()}")
    print("=" * 70)
    print(f"{'ID':<5} {'Name':<20} {'Age':<5} {'Crime':<20} {'Status':<10}")
    print("-" * 70)

    for r in results:
        print(f"{r[0]:<5} {r[1]:<20} {r[2]:<5} {r[3]:<20} {r[4]:<10}")

    print("-" * 70)
    print(f"Total matches found: {len(results)}")
    print("=" * 70)

# Test both features
print("Showing all criminals...\n")
display_all_criminals()

print("\nSearch by crime type or name:")
keyword = input("Enter search keyword: ")
display_search_results(keyword)
```
